### 3. Selecting features
I've decided to drop Date columns. It could be important in predticting corelation to number of passengers, but such column already exists. Also DepartureAirport and ArrivalAirport are included in Route column, so they can be removed. Route column is a categorical data, so it should be encoded, but if there are different values in trainig and validation datasets it'll be better to remove this column.

```python
ryanair_flight_data_validation['Route'].unique == ryanair_flight_data['Route'].unique
```

Categorical values in training and validation data differs, so encoding can cause problems with running model on validation data so I've decided to remove it

```python
ryanair_flight_data_selected_feautures = ryanair_flight_data.drop(columns=['DepartureDate', 'DepartureYear',
                                                                           'DepartureMonth', 'DepartureDay',
                                                                           'DepartureAirport', 'ArrivalAirport',
                                                                           'FlightNumber', 'Route'])
```

```python
ryanair_flight_data_selected_feautures.head()
```

### 4. Cleaning & parsing data
At this step I'll drop null values  and then parse object type to int type. In this dataset null values are presented as strings, so to use dropna method later on it must be parsed to np.nan format. For parsing Route column i'll use OneHotEncoder, beacause it is good for regression problems.

### Parsing null strings into np.nan

```python
ryanair_flight_data_selected_feautures['BagsCount'] = [(np.nan)  if value == '(null)' else value for value in \
                                                       ryanair_flight_data_selected_feautures['BagsCount']]
ryanair_flight_data_selected_feautures['FlightBagsWeight'] = [np.nan  if value == '(null)' else value for value in \
                                                              ryanair_flight_data_selected_feautures['FlightBagsWeight']]
ryanair_flight_data_selected_feautures['ActualTOW'] = [np.nan  if value == '(null)' else value for value in \
                                                       ryanair_flight_data_selected_feautures['ActualTOW']]
ryanair_flight_data_selected_feautures['FLownPassengers'] = [np.nan  if value == '(null)' else value for value in \
                                                             ryanair_flight_data_selected_feautures['FLownPassengers']]
```

### 6. Building a model.
At this step I'll try few regression models. Due to the big size of dataset, I will not use vast gridsearch methods, because of long calculation time.

```python
y = ryanair_flight_data_selected_feautures['ActualTOW']
X = ryanair_flight_data_selected_feautures.drop(columns=['ActualTOW'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

### Ridge regression

```python
from sklearn.linear_model import Ridge
pipe = Pipeline([('preprocessing', StandardScaler()), ('regressor', Ridge())])

param_grid = {
            'preprocessing': [StandardScaler(), None],
            'regressor__alpha' : range(10, 90, 10)
}

model_Ridge = GridSearchCV(pipe, param_grid, cv=5, return_train_score=True, scoring = 'neg_root_mean_squared_error', n_jobs = 1)

model_Ridge.fit(X_train, y_train)
model_Ridge.best_params_
```

```python
metrics.r2_score(y_test, model_Ridge.best_estimator_.predict(X_test))
```

### Lasso regression

```python
from sklearn.linear_model import Lasso
pipe = Pipeline([('preprocessing', StandardScaler()), ('regressor', Lasso())])

param_grid = {
            'preprocessing': [StandardScaler(), None],
            'regressor__alpha' : range(5,20,1)
}

model_Lasso = GridSearchCV(pipe, param_grid, cv=5, return_train_score=True, scoring = 'neg_root_mean_squared_error', n_jobs = 1)

model_Lasso.fit(X_train, y_train)
model_Lasso.best_params_
```

```python
metrics.r2_score(y_test, model_Lasso.best_estimator_.predict(X_test))
```

### RandomForestRegressor

```python
pipe = Pipeline([('preprocessing', StandardScaler()), ('regressor', RandomForestRegressor())])

param_grid = {
            'preprocessing': [StandardScaler(), None],
            'regressor__n_estimators': [1000],
            'regressor__max_features': ['auto'],
            'regressor__max_depth' : [15],
            'regressor__min_samples_split' : [10],
            'regressor__min_samples_leaf' : [4]
}

model_RFR = GridSearchCV(pipe, param_grid, cv=5, return_train_score=True)

model_RFR.fit(X_train, y_train)
model_RFR.best_params_
```

### 7. Model comparison

```python
models = []
models.append(('Ridge', model_Ridge.best_estimator_))
models.append(('Lasso', model_Lasso.best_estimator_))
models.append(('Random Forest Regressor', model_RFR))
```

```python
r2_score = []
root_mean_squared_error = []
mean_absolute_percentage_error = []
for name, model in models:
    r2_score.append(metrics.r2_score(y_test, model.predict(X_test)))
    root_mean_squared_error.append(metrics.mean_squared_error(y_test, model.predict(X_test), squared = False))
    mean_absolute_percentage_error.append( metrics.mean_absolute_percentage_error(y_test, model.predict(X_test)))
```

```python
model_results = {'r2_score' : r2_score,
                'root_mean_squared_error' : root_mean_squared_error,
                'mean_absolute_percentage_error' : mean_absolute_percentage_error}
results = pd.DataFrame(data=model_results)
results.insert(loc=0, column='Method', value=['Ridge', 'Lasso', 'Random Forest Regressor'])
```

```python
results
```

Random Forest Regressor was the best
